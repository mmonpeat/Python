**9. How many votes had PSOE in total?**

```python
votes = df['PSOE'].sum()
display(votes)
```

**10. How many votes had "Unidas Podemos" according to the community region? In which there were more votes (sort the list)?**

```python
votsUP = df.groupby('Comunidad')['Unidas_Podemos'].sum().reset_index()
vots = votsUP.sort_values(by='Unidas_Podemos', ascending=False)
print("Vots de Unidas_Podemos en cada comunitat: \n")
display(vots)
mes_vots = vots.iloc[0]
print("Comunitats amb més vots: \n")
display(mes_vots)
```

**11. What was the total votes of each political party according to the community region?**

```python
vots_totals_per_partit_i_comunitat = df.groupby(['Comunidad'])[['PSOE', 'PP', 'Ciudadanos', 'Unidas_Podemos', 'VOX', 'ERC', 'JxCAT']].sum().reset_index()
print("Vots totals per partit i comunitat:")
display(vots_totals_per_partit_i_comunitat)
```

**12. How may people don't voted for each community? and in total? (The abstention corresponds to the census minus the voters)**

```python
# Calcula l'abstenció per a cada comunitat
df['Abstenció'] = df['Censo_electoral'] - df['Votantes']

# Calcular l'abstenció total
abstenció_total = df['Censo_electoral'].sum() - df['Votantes'].sum()

print("Abstenció per a cada comunitat:")
display(df[['Comunidad', 'Abstenció']])

print("\nAbstenció total:", abstenció_total)
```

**13. To which community belongs the provinces with more than 1 milion of inhabitants? Sort them by `Poblacion`**

```python
pob_comunitat = df.groupby('Comunidad')['Poblacion'].sum()

# llista de comunitats amb mes de 1M
prov_1M = pob_comunitat[pob_comunitat > 100000]

sorted_communities = prov_1M.sort_values(ascending=False)

sorted_dataframe = pd.DataFrame(sorted_communities)
print("Communities with provinces having more than 1 million inhabitants sorted by Poblacion:\n")
display(sorted_dataframe)
```

```python
file_path = "./contagi_diari_ccaa_finsMaig2021.csv"
df = pd.read_csv(file_path, sep=",")
```

**2. Quants contagiats hi hagut en total?**

```python
# La primera crida a sum() suma les dades en les columnes, retornant un resultat que consisteix en la suma de cada columna per separat.
# La segona crida a sum() suma aquest resultat parcial, que és la suma de les sumes de cada columna, per obtenir la suma total de totes les dades al llarg de les columnes.
# print(df.columns)
# Estableix la columna "CCAA" com a índex (no funka)
# df = df.set_index('CCAA')
total_contagis = df.sum().sum()
print("Total de contagis en totes les comunitats:", total_contagis)
```

**3. Quin és el dia/es amb més contagiats?**

```python
#Màxim nombre de contagis en un sol dia
max_date = df.max().idxmax()
max_contagis = df.max().max()

print(f"El dia amb més contagis va ser el {max_date} amb un total de {max_contagis} contagis.")
```

**4. Fes un diccionari que reculli la suma total de contagis a cada comunitat, on la clau sigui el nom de la comunitat i el valor, els contagis totals.**

```python
sum_contagis_per_comunitat = df.sum(axis=1)

dic = dict(zip(sum_contagis_per_comunitat.index, sum_contagis_per_comunitat))

display(dic)
```

**5. Quines han estat les 3 comunitats amb més contagis?**

```python
comunitats_ordenades = sorted(dic.items(), key=lambda x: x[1], reverse=True)

# Obté les 3 comunitats amb més contagis
tres_comunitats_mes_contagis = comunitats_ordenades[:3]

print("Les 3 comunitats amb més contagis són: \n")
for comunitat, contagis in tres_comunitats_mes_contagis:
    print(f"{comunitat}: {contagis} contagis")
```

**6. Fes una llista que reculli els contàgis diàris a Catalunya.**

```python
contagis_cat = df.loc["Catalunya"]
contagis_diari_cat = contagis_cat.tolist()

print(contagis_diari_cat)
```

**7. Quina ha estat la mitjana de contagis a cada comunitat? (arrodoneix el resultat a dos decimals). Afegeix aquests valors com a nova columna que s'anomeni "Mitjana"**

```python
mitjana_contagis = df.mean(axis=1)

# Arrodoneix la mitjana a dos decimals
mitjana_contagis = mitjana_contagis.round(2)

df['Mitjana'] = mitjana_contagis
display(df)
```
