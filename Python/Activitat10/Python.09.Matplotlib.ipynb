## 9.1 Representació gràfica de funcions

La instrucció bàsica per representar gràficament una funció o una col·lecció de dades és `plt.plot()`. Aquesta funció representa gràficament una col·lecció de dades que s'introdueixen com a argument de la funció. En el següent exemple:

1. Generem un array de dades i l'assignem a una variable: `y_data`
2. Representem gràficament els valors emprant la instrucció `plt.plot()` i la variable `y_data` com a argument de la funció.

````python
y_data = np.arange(7)
print(y_data)
print("Dades: {}".format(y_data))
plt.plot(y_data)
plt.show()
````

Fixeu-vos que, en donar només el conjunt de dades, el mateix matplotlib assigna com a valor de les abcisses la posició que ocupa valor en l'array. Així al valor y=0 que es troba en primer lloc en l'array li correspòn a x=0, y=1 a x=1, etc. Tanmateix, en una representació matemàtica d'una funció o d'una col·lecció de dades experimentals, generalment disposem del valor de l'abcissa i la seva ordenada. Si introduïm dos col·leccions de dades com a arguments de la funció `plt.plot()` aleshores la primera s'utilitzarà com a abcisa (eix 'x') i la segona com a ordenada (eix 'y'):

```python
#representem gràficament els valors de dades_y en funció de dades_x
plt.plot(dades_x, dades_y)
```

Com a primera experiència amb Matplotlib generarem, pas a pas, la representació gràfica de la funció:

$$ f(x)=x^2$$



- En primer lloc, emprant la funció `numpy.linspace()` genereu un array que contingui 15 valors equiespaiats entre -2.0 i 2.0. Anomeneu aquest array "`x_data`":

````python
x_data = np.linspace(-2.0, 2.0, 15)
print(x_data)
````

+ A continuació genereu un nou array anomenat "`y_data`" que contingui el valor de la funció $x^2$ en tots els punts de l'array `x_data` anterior:

````python
y_data = x_data**2
print(y_data)
````

+ Finalment, anem a fer la representació gràfica. Emprant la instrucció `plt.plot()` representeu gràficament `y_data` en funció de `x_data`.

````python
plt.plot(x_data, y_data)
````

### **Variar els límits dels eixos**

La instrucció `plt.plot()` fa un ajust automàtic del límit de representació dels eixos. Es pot donar el cas que volguem variar manualment aquests límits per posar en evidència alguna característica de la funció que estem representant, per exemple ressaltar un màxim o ampliar una zona determinada. Si aquest és el cas, hem d'utilitzar les següents funcions que redefineix els límits dels eixos entre els valors min i max:

```python
plt.xlim(xmin, xmax) 
plt.ylim(ymin, ymax)
```

Com en el cas de les comandes per afegir títols als eixos o al gràfic, només tenen sentit dins de la mateixa cel·la on generem el gràfic.

Repetiu el gràfic anterior representant la funció en la finestra $x \in [-2.5, 2.5]$ i $y \in [-0.5, 4]$.

````python
x_data = np.linspace(-2.0, 2.0, 15)
y_data = x_data**2
plt.plot(x_data, y_data)
plt.xlim(-2.5, 2.5)
plt.ylim(-0.5, 4)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title("Representasció gràfica de x E[-2.5, 2.5] i yE[-0.5,4]")
````

### **Representar diverses funcions simultàniament**

Si en el mateix gràfic volem representar gràficament dues funcions a la vegada, només cal que en la mateixa cel·la hi escrivim dues vegades la instrucció `plt.plot()`, una per cada funció que volem representar. En la cel·la següent representeu les funcions $x^2$ i $x^4$ en un mateix gràfic: (_Suggeriment:_ Utilitzeu l'array `y_data` que conté $x^2$ per generar un nou array `y2_data` que contingui $x^4$.)

````python
x_data = np.linspace(-2.0, 2.0, 15)
y_data = x_data**2
y2_data = x_data**4
plt.plot(x_data, y_data, label='x^2')
plt.plot(x_data, y2_data, label='x^4')
plt.xlabel('x')
plt.ylabel('f(x)')
````

### **Llegendes i estils de representació**

Quan representem simultàniament diferents col·leccions de dades en una mateixa figura, és convenient representar les dades amb diferents estils per poder-les diferenciar i afegir una etiqueta que les identifiqui.

La instrucció `plt.plot()` té una sèrie d'arguments (opcionals) que permeten controlar l'estil de representació de les dades:

   - després d'especificar les dades a representar, el paràmetre `linestyle` es pot indicar l'**estil de representació**. Aquest és per defecte una línia contínua però es pot canviar per línia a traços, a punts, símbols, símbols i ratlles.
   - dins de la mateixa instrucció `plt.plot()`, es pot assignar una etiqueta a les dades representades, per això cal emprar la paraula clau: `label="text_etiqueta"`

Així podem introduir una instrucció com ara:

```python
plt.plot(x_data, y_data,'g--', label="x2")
```

On es representa `y_data` en funció de `x_data` emprant:

- una línia amb traç discontinu de color verd (_green_): `g--`
- i etiquetant la corba amb el text `x2`.
- també podem fer servir l'argument `color`

Un cop definits tots els gràfics que volem dins la figura amb tantes instruccions `plt.plot()` com calgui, cal que introduïm la instrucció `plt.legend()` per a que aparegui una finestreta en el gràfic, anomenada llegenda,  on s'especifica l'etiqueta de cada una de les corbes representades. Aquesta llegenda ve representada dins un requadre que si el volem elimina, hem d'escriure: `plt.legend(frameon=False)`. Podem canviar la posició de la llegenda amb el paràmetre `loc` que accepta entre d'altres `upper left`, `upper right`, `lower left`, `lower right`, `center` o `best`.

Afegiu una llegenda al gràfic que hem creat:

````python
x_data = np.linspace(-2.0, 2.0, 15)
y_data = x_data**2
y2_data = x_data**4
plt.plot(x_data, y_data, label='x^2')
plt.plot(x_data, y2_data, label='x^4')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(loc='upper right') # plt.legend() per defecte surt a baix a l'esquerra
````

## 9.2 Gràfics de Barres

Suposem que tenim el nombre d’alumnes aprovats i aprovades d’una assignatura en relació amb diferents cursos. 

````python
cursos = ['Programació', 'Bioinformàtica', 'Empresa', 'Marques', 'bbdd']
aprovats = [15,25,18,12,31]
plt.plot(cursos,aprovats)
plt.xlabel('Assignatures', fontsize = 15)
plt.ylabel('aprovats i aprovades', fontsize = 15)
plt.show()
````

Per canviar aquest gràfic i fer-lo de barres, només caldrà canviar la funció `plot()` per la funció `bar()` dels aprovats. Copia el codi de la cel·la anterior i fés aquesta petita variació de funcions, veuras com al executar-lo el gràfic ha canviat: 

````python
cursos = ['Programació', 'Bioinformàtica', 'Empresa', 'Marques', 'bbdd']
aprovats = [15,25,18,12,31]
plt.bar(cursos,aprovats)
plt.xlabel('Assignatures', fontsize = 15)
plt.ylabel('aprovats i aprovades', fontsize = 15)
plt.show()
````

La funció `bar()` permet introduir més paràmetres, per tal de personalitzar el disseny del gràfic. Per exemple, podem modificar la funció, escollint el color de les barres i la seva amplada. Es realitza amb els paràmetres `color` i `width`

````python
plt.bar(cursos,aprovats, color='red', width=0.25)
````

També tenim l’opció d’afegir un color a l’entorn de la barra (`edgecolor`) i donar-li un gruix (`linewidth`)

````python
plt.bar(cursos,aprovats, color='red', width=0.5, edgecolor='purple', linewidth=2.0) 
````

Si, en lloc de treballar amb un gràfic de barres vertical, volem que sigui horitzontal, haurem de canviar de funció. En aquest cas, la funció serà `barh()`.

````python
plt.barh(cursos,aprovats, color='red', height = 0.5, edgecolor='green', linewidth=2.0) 
````

Fixem-nos que els paràmetres de les dues funcions: `bar()` i `barh()` són pràcticament els mateixos, tret del paràmetre `width` de `bar()`, que, en la funció `barh()`, passa a ser `height`. **ATENCIÓ ! També hem d'intercanviar els valors de xlabel i ylabel, per ser coherents amb la nova orientació.**

## 9.3 Histogrames

Un histograma és una representació gràfica associada a una variable estadística contínua (o discreta amb molts valors diferents) per visualitzar les freqüències de les dades. Una variable contínua seria per exemple el pes o l'alçada que pot prendre molts valors. Es construeix a partir d'una taula de freqüències o distribució de freqüències amb les dades agrupades en classes. 

Veiem un exemple senzill de com crear un histograma a Matplotlib, creant un array random 


````python
x = np.random.randn(10000) # Normally distributed random numbers
info = plt.hist(x, 20, facecolor='g', edgecolor="b", alpha=0.75) #Alpha és un valor 0-1 que fa referència a la transparència de les barres
plt.grid(True)
plt.xlabel('Variable')
plt.ylabel('Probabilitat')
plt.title('Histograma')
````

## **Exercici 1**

Una mesura de l'obesitat es determina mitjançant l'índex de massa corporal (IMC), que es calcula dividint els kilograms de pes pel quadrat de l'alçada en metres:

```python
     IMC = peso (kg)/altura (m) ^2
```
De manera que si:
+ IMC<=19: baix pes
+ IMC<=25: pes correcte
+ IMC<=30: sobrepes
+ IMC>30: obesitat

**1. Genera una mostra de 20 persones amb les dades de pes i altura que siguin randoms i dins d'un marge de realitat.**

````python
pes_min, pes_max = 45, 120
alçada_min, alçada_max = 1.50, 2.10
pesos = np.random.uniform(pes_min, pes_max, 20)
alçades = np.random.uniform(alçada_min, alçada_max, 20)
for i in range(20):
    print(f"Persona {i + 1}: Pes = {pesos[i]:.2f} kg, Alçada = {alçades[i]:.2f} m")
````

**2. Calcula el IMC.**

````python
IMC = pesos / (alçades**2)
IMC
````

**3. Dibuixa un gràfic de dispersió que relacioni l'altura (x) i el pes (y). Recorda posar noms als eixos, títols i llegenda**

````python
pes_min, pes_max = 45, 120
alçada_min, alçada_max = 1.50, 2.10
pesos = np.random.uniform(pes_min, pes_max, 20)
alçades = np.random.uniform(alçada_min, alçada_max, 20)
IMC = pesos / (alçades**2)
plt.scatter(alçades, pesos, c=IMC, cmap='viridis', marker='o', label='Persones')
plt.xlabel('Alçada (m)')
plt.ylabel('Pes(kg)')
plt.title('Relació entre Alçada i Pes')
plt.legend()
plt.colorbar(label='IMC')
plt.show()
````

**4. Dibuixa un histograma que reculli els valors de IMC**

````python
pes_min, pes_max = 45, 120  # En quilograms
altura_min, altura_max = 1.50, 2.10  # En metres
pesos = np.random.uniform(pes_min, pes_max, 20)
alçades = np.random.uniform(altura_min, altura_max, 20)
imc = pesos / (alçades ** 2)
plt.hist(imc, bins=10, color='skyblue', edgecolor='black')
plt.xlabel('IMC')
plt.ylabel('Nombre de persones')
plt.title('Distribució de l\'IMC')
plt.show()
````

**5. Coloreja a l'histograma les barres en funció dels criteris de IMC i inclou-ho a la llegenda**

````python
colors = ['lightgreen', 'lightblue', 'orange', 'red']
rangos_IMC = [0, 19, 25, 30, np.inf]

plt.figure(figsize=(8, 6))
n, bins, patches = plt.hist(imc, bins=10, color='blue', edgecolor='black', alpha=0.7)

for j in range(len(patches)):
    for i in range(len(rangos_IMC) - 1):
        if bins[j] >= rangos_IMC[i] and bins[j] < rangos_IMC[i + 1]:
            patches[j].set_fc(colors[i])

plt.xlabel('Índice de Masa Corporal (IMC)')
plt.ylabel('Frecuencia')
plt.title('Distribución del IMC')
plt.grid(True, linestyle='--', alpha=0.5)

handles = [plt.Rectangle((0, 0), 1, 1, color=colors[i], ec="k") for i in range(len(colors))]
labels = ['Baix Pes', 'Pes Correcte', 'Sobrepes', 'Obesitat']
plt.legend(handles, labels)

plt.show()
````
