**4. Fes un diccionari que reculli la suma total de contagis a cada comunitat, on la clau sigui el nom de la comunitat i el valor, els contagis totals.**

````python
contagis = df.sum(axis=1).to_dict()
contagis
````

**5. Quines han estat les 3 comunitats amb més contagis?**

````python
maxcontagis = sorted(list(contagis.values()),reverse=True)[:3]
````

````python
for ccaa in contagis:
    if contagis[ccaa] in maxcontagis:
        print(ccaa)
````

**7. Quina ha estat la mitjana de contagis a cada comunitat? (arrodoneix el resultat a dos decimals). Afegeix aquests valors com a nova columna que s'anomeni "Mitjana"**

````python
df["Mitjana"] = df.mean(axis=1).round(2)
df["Mitjana"]
````

````python
# Opció 2
def mitjana(row):
    mitjana = row.sum() / row.shape[0]
    mitjana = "%0.2f" % mitjana
    return mitjana

df.apply(mitjana, axis=1)
````
